# Imports, Loads and Utils

```python
#Import
import numpy as np
import  pandas as pd
import seaborn as sns
import matplotlib.pyplot as mpl
```

```python
cleanup = {}
#-utility function
def create_enum(name, *kwargs):
    global cleanup
    inner = {}
    for (ind, val) in enumerate(kwargs):
        inner[val] = ind
    cleanup[name] = inner
```

```python
#---------------loading
file = pd.read_csv("titanic-passengers.csv", sep = ";")
filecopy = file.copy()

file.head()
```

# PREPROCESSING

```python
#----------------Cleanup phase

create_enum("Survived", "Yes", "No")
create_enum("Embarked", 'S', 'C', 'Q')
create_enum("Sex", "male", "female")



file['Age'].fillna(file['Age'].mean(), inplace=True) #the mean (probably) shouldn't change the output

file['Embarked'].fillna(value ='S', inplace=True) # Used the highest used value since there is only two
#missing values and S is already dominating the set
try:
    file.drop("Cabin", axis = 1, inplace=True)         # too much missing data
    file.drop("Ticket", axis = 1, inplace=True)        # irrelevant
    file.drop("PassengerId",axis = 1, inplace=True)    # irrelevant 
except Exception as e:
    print(f"Error trying to delete rows\n.Try refreshing the loading part of the program.\n{e}")
    
file.replace(cleanup, inplace= True)

file.head()


```

```python
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = mpl.subplots( figsize =( 12 , 10 ))
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={'shrink' : .9}, 
        ax=ax, 
        annot = True, 
        annot_kws = {'fontsize' : 12})

    
plot_correlation_map(file)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
r = list(file.columns)
r.remove("Survived")
x = file[r]
y = file['Survived']

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25, random_state= 0)
```

```python
logreg = LogisticRegression()   
logreg.fit(x_train, y_train)  
y_pred  = logreg.predict(x_test)    
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```
