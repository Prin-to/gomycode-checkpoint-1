```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
rf = pd.read_csv("Market_Basket_Optimisation.csv")

te = TransactionEncoder()
te_ary = te.fit(rf).transform(rf) 
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
print(df["a"].unique())
df.replace( {df["a"].unique()[0]: 0, df["a"].unique()[1] : 1}, inplace = True)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True) 

frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
association_rules(frequent_itemsets, metric="lift",min_threshold=1.25)
```
